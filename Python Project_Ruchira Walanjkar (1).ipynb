```python
#EDA
#Pie chart for Gender Distribution 
custom_palette=sns.color_palette("Set2")

plt.figure(figsize=(10,5))
df['Gender'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=custom_palette)
plt.title('Gender_Destribution')
plt.show()
```

```python
##EDA
#Pie chart for Preferred_Category 
custom_palette=sns.color_palette("Set1")

plt.figure(figsize=(10,7))
df['Preferred_Category'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=custom_palette)
plt.title('Categorywise_Destribution')
plt.show()
```

```python
##Histogram for Age Distribution
sns.histplot(data=df,x='Age',kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
##Histogram for Annual Income
sns.histplot(data=df['Annual_Income'],kde=True)
plt.title('Distribution of Annual Income')
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.show()
```

```python
##Countplot for Marketing_Channel by Gender
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Marketing_Channel", hue="Gender", palette="pastel")
plt.title("Marketing Channel Distribution by Gender")
plt.xlabel("Marketing Channel")
plt.ylabel("Count")
plt.show()
```

```python
#Countplot for  Payment_Method by gender
sns.countplot(data=df, x='Payment_Method', hue= 'Gender',palette='pastel')
```

```python
#Preferrd category by gender
sns.countplot(data=df, x='Preferred_Category', hue='Gender', palette='Set2')
```

```python
#Countplot of Payment Method by City (only top 5)
#Step 1: Find top 5 cities by total spending
top5_cities = df.groupby("City")["Total_Spending"].sum().nlargest(5).index

# Step 2: Filter dataset for those cities
df_top5 = df[df["City"].isin(top5_cities)]

# Step 3: Countplot of Payment Method by City (only top 5)
plt.figure(figsize=(12,6))
sns.countplot(data=df_top5, x="Payment_Method", hue="City", palette="Set3")
plt.title("Payment Method Distribution in Top 5 Cities by Spending")
plt.xlabel("Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
#Histogram for Purchase_Frequency
sns.histplot(data=df, x='Purchase_Frequency', kde=True)
plt.title('Distribution by Purchase_Frequency')
```

```python
##Scatterplot for  and Age

sns.scatterplot(data=df_top5,x='Total_Spending', y='Age')
plt.title('Total_Speinding and Age')
sns.regplot(data=df_top5, x='Total_Spending', y='Age', scatter=False, color="red", line_kws={"linewidth":2})
plt.title("Total Spending vs Age (with Trend Line)")
plt.xlabel("Total Spending")
plt.ylabel("Age")
plt.show()
```

```python
##Boxplot and Violinplot for Preferred Category and Satisfaction Score
plt.figure(figsize=(16,10))

plt.subplot(1,2,1)
sns.boxplot(data=df, x='Preferred_Category', y='Satisfaction_Score')
plt.title('Preferred_Category vs Satisfaction_Score(Box_Plot)')

plt.subplot(1,2,2)
sns.violinplot(data=df, x='Preferred_Category', y='Satisfaction_Score')
plt.title('Preferred_Category vs Satisfaction_Score(Violinplot)')
```

```python
#Boxplot for Purchase_frequency Vs Marketing_channel
plt.figure(figsize=(16,10))

plt.subplot(1,2,1)
sns.boxplot(data=df, x='Marketing_Channel', y='Purchase_Frequency')
plt.title('Marketing_Channel vs Purchase_Frequency(Box_Plot)')

plt.subplot(1,2,2)
sns.violinplot(data=df, x='Marketing_Channel', y='Purchase_Frequency')
plt.title('Marketing_Channel vs Purchase_Frequency(Violinplot)')


```

```python
#Barplot for Annual Income and City
sns.barplot(data=df, x='City', y='Annual_Income')
plt.xticks(rotation=90)
```

```python
#Barplot for Preferred_category and Avg_Order_Value
sns.barplot(data=df, x='Preferred_Category', y='Avg_Order_Value')
plt.title('Category wise Avg_OrderValue')
plt.xticks(rotation=90)
```

```python
import matplotlib.pyplot as plt

# Aggregate average order value per category
category_avg = df.groupby("Preferred_Category")["Avg_Order_Value"].mean()

plt.figure(figsize=(10,6))
plt.bar(category_avg.index, category_avg.values, color="skyblue")
plt.title("Category-wise Avg Order Value")
plt.xlabel("Preferred Category")
plt.ylabel("Avg Order Value")
plt.xticks(rotation=90)
plt.show()
```

```python
#Category_wise order value(Horizontal))
plt.figure(figsize=(10,6))
sns.barplot(data=df, y='Preferred_Category', x='Avg_Order_Value',hue='Preferred_Category',palette="Set3")
plt.title('Category-wise Avg Order Value(Horizontal)')
plt.xlabel("Avg Order Value")
plt.ylabel("Preferred Category")
plt.show()
```
